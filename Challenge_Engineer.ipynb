```python
import json
import datetime
import pandas as pd
from urllib import response
import requests
```

Conexion al enlace y exploración

```python
url ='https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow'
response = requests.get(url).json()
data=response["items"]
df=pd.json_normalize(data,max_level=1)
df.head()
```

Obtener el número de respuestas contestadas y no contestadas

```python
respondidas = "Total de preguntas respondidas: " + str( df[df["is_answered"]==True]["is_answered"].count())
no_respondidas = "Total de preguntas no respondidas: " + str( df[df["is_answered"]==False]["is_answered"].count())
print(respondidas)
print(no_respondidas)
```

Obtener la respuesta con menor número de vistas

```python
df[df["view_count"]==df["view_count"].min()]
```

Obtener la respuesta más vieja y más actual

```python
## Funcion para convertir la fecha en un formato correcto
def parseDate(date):
     fmt = "%Y-%m-%d %H:%M:%S"
     t_utc = datetime.datetime.utcfromtimestamp(float(date)/1000.)
     return t_utc.strftime(fmt)
```

```python
df["creation_date"]=df["creation_date"].apply(parseDate)
df["last_activity_date"]=df["last_activity_date"].apply(parseDate)
```

```python
df[df["creation_date"]==df["creation_date"].min()]
```

```python
df[df["creation_date"]==df["creation_date"].max()]
```

Obtener la respuesta del owner que tenga una mayor reputación

```python
df[df["owner.reputation"]==df["owner.reputation"].max()]
```
